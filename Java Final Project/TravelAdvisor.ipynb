```python
import pandas as pd
data_path = "google_review_ratings.csv"
```

```python
#read the data
data = pd.read_csv(data_path)
data = data.drop(columns = ["Unnamed: 25"])
data = data.dropna()
data.head()
```

```python
#create X and y. y is just monument ratings, X is all other attraction ratings
X = data.drop(columns = ["Category 24", "User"])
y = data["Category 24"]

#split the data in to training and testing data at 80:20 ratio
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#fit a linear regression model to the data
model.fit(X_train, y_train)
```

```python
#make predictions for all users in the test set
preds = model.predict(X_test)
preds
```

```python
y_test = list(y_test)
```

```python
#calculate average error
ans = 0
for i in range (len(y_test)):
    print(abs(y_test[i] - preds[i]))
    ans += abs(y_test[i] - preds[i])

ans/len(preds)
```

```python
#create a model to compare to the regression model, which just makes its prediction the average of all other ratings
avgs = X_test.sum(axis=1)
avgs = avgs/23
avgs = list(avgs)
```

```python
#calculate average error for this second model
ans = 0
for i in range (len(y_test)):
    print(abs(y_test[i] - avgs[i]))
    ans += abs(y_test[i] - avgs[i])

ans/len(avgs)
```
